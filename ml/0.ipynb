# 多元线性模型

## 建立模型

```python
def computeCost(X, y, theta):
    inner = np.power(((X * theta.T) - y), 2)
    return np.sum(inner) / (2 * len(X))


def gradientDescent(X, y, theta, alpha, iters):
    temp = np.matrix(np.zeros(theta.shape))
    parameters = int(theta.ravel().shape[1])
    cost = np.zeros(iters)

    for i in range(iters):
        error = (X * theta.T) - y

        for j in range(parameters):
            term = np.multiply(error, X[:, j])
            temp[0, j] = theta[0, j] - ((alpha / len(X)) * np.sum(term))

        theta = temp
        cost[i] = computeCost(X, y, theta)

    return theta, cost
```

```python
x = train[tezheng]
x.insert(0, 'Ones', 1)

x = np.matrix(x.values)
y = np.matrix(train['price'].values).T
theta = np.matrix(np.zeros((1, x.shape[1])))
```

```python
train_test_x = train[tezheng + ['price']].sample(frac=0.5).copy()
```

## 交叉验证和模型评估

```python
alpha = 0.01
iters = 1000
a = 0.01
MAE = float('inf')
for j in range(60):

    theta = np.matrix(np.zeros((1, x.shape[1])))
    g, cost = gradientDescent(x, y, theta, alpha, iters)

    train_test_y = train_test_x.copy()
    train_test_y['price'] = 0
    for i in range(len(tezheng)):
        train_test_y['price'] += g[0, i + 1] * train_test_y[tezheng[i]]
    train_test_y['price'] += g[0, 0]
    train_test_y['price'] = train_test_y['price'] * train_dif + train_min

    train_test_y['price'] = np.expm1(train_test_y['price'])

    yhat = np.expm1(train_test_x.price * train_dif + train_min)

    dif = (yhat - train_test_y['price']).abs().sum() / \
        train_test_y['price'].count()

    if dif < MAE:
        MAE = dif
        a = alpha

    alpha += 0.01

    print('第%d次迭代' % j, dif)
```
